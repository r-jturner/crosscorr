```python
import numpy as np
import matplotlib.pyplot as plt
import time
import corr_desi_wrapper as corr
```

```python
dat = np.load("./dat/desi_dat.npy")
ran = np.load("./dat/desi_ran.npy")
```

```python
start = time.time()

DDn,DDd = corr.corrPairCount(dat, dat, 125, 5, "psi3", "u", 2)
DRn,DRd = corr.corrPairCount(dat, ran, 125, 5, "psi3", "u", 2)
RDn,RDd = corr.corrPairCount(ran, dat, 125, 5, "psi3", "u", 2)
RRn,RRd = corr.corrPairCount(ran, ran, 125, 5, "psi3", "u", 2)

end = time.time()
print("Total time to run is",np.round(end-start,3),"seconds.")
```

```python
desi_psi3 = corr.calc_psi3(10000,50000,DDn,DRn,RDn,RRn,RRd)
plt.plot(np.arange(0,125,5), np.arange(0,125,5) * desi_psi3, marker="o",color="black")
```
